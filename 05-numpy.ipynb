### Vector calculation

```python
size = 10
xs1 = np.random.rand(size)
xs2 = np.ones(size)

print(xs1)
print(xs2)
xs1 * xs2
```

```python
np.dot(xs1, xs2) # product
```

```python
dot = 0
for k in range(size):
    dot += xs1[k] * xs2[k]
dot
```
